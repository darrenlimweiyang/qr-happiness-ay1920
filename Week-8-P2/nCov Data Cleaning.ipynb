```python
import pandas as pd
```

```python
df = pd.read_csv("covid_19_data.csv")
```

```python
df.columns.tolist()
```

```python
df.drop(columns='SNo', axis=1,
       inplace=True)
```

```python
df['Province/State'].fillna(" ",
                           inplace=True)
```

```python
df.drop(columns=['Recovered',
                'Deaths',
                'ObservationDate'], axis=1,
       inplace=True)
```

```python
df['Combined'] = df['Country/Region'] + " " + df['Province/State']
```

```python
df.drop(columns=['Province/State',
                'Country/Region'], axis=1,
       inplace=True)
```

```python
df['Test'] = df['Combined'] + df['Last Update']
```

```python
a = df['Test'].value_counts().reset_index()

b = a[a['Test'] == 1]
```

```python
country_list = b['index'].tolist()
```

```python
df = df[df['Test'].isin(country_list)]
```

```python
pivoted = df.pivot(index='Combined', 
        columns='Last Update',
        values='Confirmed')
```

```python
final_df = pivoted.reset_index()
```

```python
final_df.head(3)
```

```python
final_df['Country'] = final_df['Combined'].apply(lambda x: x.strip())
```
