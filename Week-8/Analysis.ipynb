```python
import pandas as pd
import seaborn as sns
import re

sns.set(rc={'figure.figsize':(20,15)})

sns.set_style("whitegrid")

import matplotlib.pylab as plt
```

```python
df = pd.read_csv("Covid-Dataset.csv")
```

```python
df.head(2)
```

```python
new_df = df.drop(columns=['Case ',
                 'Status',
                 'Symtomatic at',
                 'Displayed Symptoms',
                 'Recovered At',
                 'Days to recover',
                 'Symptomatic to Confirmation'
                ], axis=1)
```

```python
from datetime import datetime

def test_func(x):
    datetime_object = datetime.strptime(x, '%d %b %Y')
    return datetime_object
    
test_func("9 Mar 2020")
```

```python
new_df.dropna(inplace=True)
```

```python
new_df['Dummy Date'] = new_df['Confirmed At'].apply(lambda x:x.split(","))
new_df['Dummy Day'] = new_df['Dummy Date'].apply(lambda x:x[0])

new_df['Dummy Day Int Only'] = new_df['Dummy Day'].apply(lambda x:re.search(r'\d+',x).group())

new_df['Date String'] = new_df['Dummy Day Int Only'] + new_df['Dummy Date'].apply(lambda x:x[1])

new_df['Datetime Object'] = new_df['Date String'].apply(test_func)

```

```python
new_df.drop(columns=['Confirmed At',
                    'Dummy Date',
                    'Dummy Day',
                    'Dummy Day Int Only',
                    'Date String'],
           axis=1,inplace=True)
```

```python
new_df.groupby('Datetime Object').count().cumsum()['Patient'].reset_index().head(2)
```

```python
sns.lineplot(x='Datetime Object',
             y='Patient',
             data=new_df.groupby('Datetime Object').count().cumsum()['Patient'].reset_index())
```

```python
sns.barplot(x='Datetime Object',
             y='Patient',
             data=new_df.groupby('Datetime Object').count().cumsum()['Patient'].reset_index() ,
           color='salmon')
```

```python
new_df_recovered = df.drop(columns=['Status',
                'Case ',
                'Symptomatic to Confirmation',
                'Days to recover',
                'Symtomatic at',
                'Displayed Symptoms'],
       axis=1)
```

```python
new_df_recovered.dropna(inplace=True)
```

```python
new_df_recovered = new_df_recovered[new_df_recovered['Recovered At'] != "-"]
```

```python
new_df_recovered['Dummy Date'] = new_df_recovered['Recovered At'].apply(lambda x:x.split(","))
new_df_recovered['Dummy Day'] = new_df_recovered['Dummy Date'].apply(lambda x:x[0])

new_df_recovered['Dummy Day Int Only'] = new_df_recovered['Dummy Day'].apply(lambda x:re.search(r'\d+',x).group())
 
new_df_recovered['Date String'] = new_df_recovered['Dummy Day Int Only'] + new_df_recovered['Dummy Date'].apply(lambda x:x[1])

new_df_recovered['Datetime Object'] = new_df_recovered['Date String'].apply(test_func)
```

```python
new_df_recovered.drop(columns=['Dummy Date',
                              'Dummy Day Int Only',
                              'Date String',
                              'Dummy Day'],
                     inplace=True)
```

```python
cumsum_daily_confirmed = new_df.groupby('Datetime Object').count().cumsum()['Patient'].reset_index() 
```

```python
cumsum_daily_recovered = new_df_recovered.groupby('Datetime Object').count().cumsum()['Age'].reset_index() 
```

```python
combined_df = pd.merge(cumsum_daily_confirmed,
                      cumsum_daily_recovered,
                      how='left',
                      on='Datetime Object')
```

```python
combined_df.rename(columns={'Patient':'Confirmed Case Count',
                           'Age': 'Recovered Case Count'},
                  inplace=True)
```

```python
ax = sns.lineplot(x='Datetime Object',
             y='value',
             hue='Type',
             data=pd.melt(combined_df,
       id_vars=['Datetime Object'],
       value_vars=['Confirmed Case Count','Recovered Case Count']).rename(columns={'variable':'Type'}),
                 legend='full')

ax.axes.set_title("Singapore: Daily Cases: Confirmed VS Recovered",fontsize=25)
ax.set_xlabel("Date",fontsize=20)
ax.set_ylabel("Number of cases",fontsize=15)
ax.tick_params(labelsize=20)

ax.legend(fontsize='x-large', title_fontsize='40')
```
